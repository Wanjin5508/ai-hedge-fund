```python
import yfinance as yf
import matplotlib.pyplot as plt

# 获取SPDR黄金ETF（GLD）过去4个月数据
gld = yf.Ticker("GLD")
hist = gld.history(period="4mo")

# 绘图
hist['Close'].plot(title='GLD (黄金ETF)近4个月价格走势')
plt.xlabel("日期")
plt.ylabel("收盘价 (USD)")

plt.grid(True)
plt.show()

```
